# JUEGO DEL AHORCADO

- Reglas del juego del "Ahorcado":

    - Un jugador elige una palabra secreta y dibuja un espacio para cada letra de la palabra.

    - El otro jugador intenta adivinar letras para completar la palabra.

    - Si el jugador adivinador adivina una letra correctamente, se revela en su lugar correspondiente.

    - Si el jugador adivinador adivina incorrectamente, se dibuja una parte del cuerpo en la horca.

    - El objetivo del jugador adivinador es adivinar la palabra antes de que se dibuje el dibujo completo en la horca.

```python
#INPUT PALABRA, DIVIDIR CADA CARACTER EN LISTA, CREAR LÑISTA CON EL MISMO NUMERO DE ESPACIOS (?)
# validacion minuscula, sin espacios isalpha (sin simbolos)
# clase con self vidas self dibujos
#input jugador letras de 1 en 1 len ==1 / for enumerate /while playing vidas = 0 
```

```python
class JuegoAhorcado: 
    def __init__(self):
        self.palabra_secreta = ""
        self.espacios = []
        self.vidas = 8
        self.letras_usadas = []
    
    def definir_palabra(self):
        while True:
            entrada_provisional = input("Introduce palabra secreta.").lower().strip()
            if entrada_provisional.isalpha():
                self.palabra_secreta = entrada_provisional
                self.espacios = ["_"] * len(self.palabra_secreta)
                break
            else: 
                print ("Palabra no válida. Por favor introduce solo letras sin espacios o símbolos.")
    
    def pedir_letra(self):
        while True:
            letra = input ("Introduce una letra.").lower().strip()
            if len (letra) != 1:
                print ("Por favor, introduce una sola letra.")
            elif not letra.isalpha():
                print ("Por favor, introduce un caracter válido. Solo letras.")
            elif letra in self.letras_usadas:
                print ("¡Letra ya usada!. Sigue probando.")
            else:
                self.letras_usadas.append(letra)
                return letra
            
```

```python
Jugador1 = JuegoAhorcado()
```

```python
Jugador1.definir_palabra()
```

```python
print(f"Palabra secreta guardada: {Jugador1.palabra_secreta}")
print(f"Lista de espacios generada: {Jugador1.espacios}")
print(f"Vidas iniciales: {Jugador1.vidas}")
```

```python
player_X = "X"
player_O = "O"
current_player = player_X
board = [[" "] * 3 for _ in range(3)]
game_over = False


def mostrar_tablero():
    print("\n   1   2   3")
    for i in range(3):
        print(f"{i + 1}  {board[i][0]} | {board[i][1]} | {board[i][2]}")
        if i < 2:
            print("  ---+---+---")
    print()


def cambiar_turno():
    global current_player
    if current_player == player_X:
        current_player = player_O
    else:
        current_player = player_X


def comprobar_ganador():
    for fila in range(3):
        if board[fila][0] == board[fila][1] == board[fila][2] != " ":
            return board[fila][0]

    for columna in range(3):
        if board[0][columna] == board[1][columna] == board[2][columna] != " ":
            return board[0][columna]

    if board[0][0] == board[1][1] == board[2][2] != " ":
        return board[0][0]

    if board[0][2] == board[1][1] == board[2][0] != " ":
        return board[0][2]

    return None


def comprobar_empate():
    for fila in range(3):
        for columna in range(3):
            if board[fila][columna] == " ":
                return False
    return True


def hacer_jugada():
    global game_over

    while True:
        try:
            fila = int(input(f"Turno de {current_player}. Elige fila (1-3): ")) - 1
            columna = int(input(f"Turno de {current_player}. Elige columna (1-3): ")) - 1

            if fila < 0 or fila > 2 or columna < 0 or columna > 2:
                print("Posición fuera de rango. Prueba otra vez.\n")
                continue

            if board[fila][columna] != " ":
                print("Esa casilla ya está ocupada. Elige otra.\n")
                continue

            board[fila][columna] = current_player
            break

        except ValueError:
            print("Debes escribir un número del 1 al 3.\n")

    ganador = comprobar_ganador()

    if ganador:
        mostrar_tablero()
        print(f"👑 ¡{ganador} reina el tablero! 💃")
        game_over = True
    elif comprobar_empate():
        mostrar_tablero()
        print("🤝 ¡Nadie gana, pero qué batalla!")
        game_over = True
    else:
        cambiar_turno()


def reiniciar_juego():
    global current_player, board, game_over
    current_player = player_X
    board = [[" "] * 3 for _ in range(3)]
    game_over = False


def jugar():
    print("🎮 Bienvenida al Tres en Raya 🎮")

    while True:
        while not game_over:
            mostrar_tablero()
            hacer_jugada()

        respuesta = input("\n¿Quieres jugar otra vez? (s/n): ").lower()

        if respuesta == "s":
            reiniciar_juego()
        else:
            print("Gracias por jugar. ¡Hasta la próxima!")
            break


jugar()
```

# Tic Tac Toe

- Reglas del Juego "Tic-Tac-Toe":

    - El juego se juega en un tablero de 3x3.

    - Dos jugadores, uno representa "X" y el otro "O", alternan turnos.

    - El objetivo es obtener tres símbolos del mismo tipo en línea horizontal, vertical o diagonal.

    - El juego termina cuando un jugador gana o el tablero se llena (empate).

    - El jugador que logre colocar tres de sus símbolos en línea es el ganador.

```python
import random
```

```python
class TicTacToe:
    def __init__(self):
        # Filas con huecos
        fila0 = [" ", " ", " "]
        fila1 = [" ", " ", " "]
        fila2 = [" ", " ", " "]
        
        # Unimos las 3 filas en una lista
        self.tablero = [fila0, fila1, fila2]
        self.turno = "X"

        self.posiciones = {
            1: (0, 0), 2: (0, 1), 3: (0, 2),
            4: (1, 0), 5: (1, 1), 6: (1, 2),
            7: (2, 0), 8: (2, 1), 9: (2, 2)
        }

    # Mostramos el tablero por índices de lista y posición:
    def mostrar_tablero(self):
        print(self.tablero[0][0], " |", self.tablero[0][1], "| ", self.tablero[0][2])
        print("---+---+---")
        print(self.tablero[1][0], " |", self.tablero[1][1], "| ", self.tablero[1][2])
        print("---+---+---")
        print(self.tablero[2][0], " |", self.tablero[2][1], "| ", self.tablero[2][2])

    # ¿Alguien ha ganado ya? Combinaciones ganadoras:
    def verificar_ganador(self, jugador):
        t = self.tablero
        # Por Filas
        if (t[0][0] == t[0][1] == t[0][2] == jugador) or \
           (t[1][0] == t[1][1] == t[1][2] == jugador) or \
           (t[2][0] == t[2][1] == t[2][2] == jugador):
            return True
        
        # Por Columnas
        if (t[0][0] == t[1][0] == t[2][0] == jugador) or \
           (t[0][1] == t[1][1] == t[2][1] == jugador) or \
           (t[0][2] == t[1][2] == t[2][2] == jugador):
            return True
            
        # Por Diagonales
        if (t[0][0] == t[1][1] == t[2][2] == jugador) or \
           (t[0][2] == t[1][1] == t[2][0] == jugador):
            return True
            
        return False



    #Chulta de posiciones
    def mostrar_guia(self):
        print("\n--- GUÍA DE POSICIONES ---")
        print(" 1 | 2 | 3 ")
        print("---+---+---")
        print(" 4 | 5 | 6 ")
        print("---+---+---")
        print(" 7 | 8 | 9 ")
        print("--------------------------\n")

    # Funcion del juego:
    def jugar(self):
        self.mostrar_guia()

        modo = input("¿Quieres jugar contra un amigo (1) o contra la máquina (2)?: ")
        if modo == "2": 
            self.contra_pc = True
        else: 
            self.contra_pc = False
    

        print("¡Empieza el Tres en Raya!")
        movimientos = 0
        
        while movimientos < 9:
            self.mostrar_tablero()
            print("Turno de:", self.turno)
            
            #No sabe elegir el mejor lugar
            if self.contra_pc and self.turno == "O":
                print("La máquina está eligiendo...")
                eleccion = random.randint(1, 9)
                fila, col = self.posiciones[eleccion]
                if self.tablero[fila][col] != " ":
                    continue 
            
            else:
                try:
                    eleccion = int(input("Elige casilla (1-9) o elige <0> para salir: "))
                    if eleccion == 0:
                        print("\n🚫 Juego cancelado.")
                        return
                    fila, col = self.posiciones[eleccion]
                except:
                    print("\n❌ ERROR: Introduce un número del 1 al 9.")
                    continue

            #¿Está vacío?
            if self.tablero[fila][col] == " ":
                self.tablero[fila][col] = self.turno
                movimientos = movimientos + 1
                
                #Comprobar si alguien gana
                if self.verificar_ganador(self.turno):
                    self.mostrar_tablero()
                    print("¡Ganó el jugador", self.turno, "!")
                    return # Se acaba el juego
                
                # A quien le toca
                if self.turno == "X":
                    self.turno = "O"
                else:
                    self.turno = "X"
            else:
                print("¡Esa casilla ya está llena!")
        
        self.mostrar_tablero()
        print("¡Empate!")
```

```python
# Ejecutar
juego = TicTacToe()
juego.jugar()
```

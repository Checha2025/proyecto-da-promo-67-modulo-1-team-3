```python
# módulo para generar aleatorios
import random 

# tupla que contiene las preguntas
questions = (
    "¿Cuál es el desierto más grande del mundo?",
    "¿Cuántas comunidades autónomas tiene España?",
    "¿Qué país es conocido como la tierra de los canguros?",
    "¿Qué río pasa por París?",
    "¿En qué país hay un mayor número de pirámides?",
    "¿Qué comunidad autónoma en España tiene el euskera como lengua oficial?",
    "¿Cuál es el lago más grande del mundo?",
    "¿Qué continente se encuentra en los 4 hemisferios?",
    "¿Rusia pertenece a Europa o a Asia?")

# tupla de tuplas, es decir tupla grande que contiene las opciones de las preguntas
options = (("Sahara","Atacama","Kalahari","Antártico"),
           ("20","19","17","24"),
           ("Nueva Zelanda","Australia","Pakistán","Grecia"),
           ("Danubio","Tajo","Ebro","Sena"),
           ("China","Egipto","Sudán","México"),
           ("País Vasco","Valencia","Galicia","Cataluña"),
           ("Victoria","Misisipi","Baikal","Mar Caspio"),
           ("América","Africa","Asia","Oceanía"),
           ("Asia","Oceanía","Europa","Antártida"))

# otra tupla para contener las respuestas
answers = (
    "Antártico",
    "17",
    "Australia",
    "Sena",
    "Sudán",
    "País Vasco",
    "Mar Caspio",
    "Africa",
    "Europa"
)

#el quiz para unir grupos elemento por elemento y que puede asi convertirse en una lista normal
quiz = list(zip(questions, options, answers))
random.shuffle(quiz) # el random.shuffle mezcla el orden de la lista, asi las preguntas no salen en el mismo orden


guesses = []   # respuestas elegidas 
score = 0      # cuenta las respuestas correctas
fails = 0      # cuenta los fallos de las respuestas

valid_options = ("A", "B", "C", "D")  # para comprobar que lo que escribe la jugadora es válido
letters = ("A", "B", "C", "D")        # para mostrar las opciones con letras

    
for question, options, answer in quiz:    # repasa cada elemento del juego 
        print("\n______________________________")    
        print(question)
    
        shuffled_options = list(options)
        random.shuffle(shuffled_options)              # el random.shuffle convierte la lista en una lista para poder mezclar las opciones
       
        for i, option in enumerate(shuffled_options):   # el enumerate recorre la lista 
            print(f"{letters[i]}. {option}")
        
        guess = input("Escribe tu respuesta (A, B, C, D): ").upper()   # input es para que la jugadora escriba algo
   
    
        while guess not in valid_options:                               # valida la respuesta y si algo es incorrecto sigue preguntando
          guess = input("Respuesta no válida. Escribe A, B, C, D: ").upper()
          
    
        selected_index = letters.index(guess)                  #selected_index busca en qué posición está la letra elegida
        selected_option = shuffled_options[selected_index]     # usar un índice para saber cuá es el texto real de la respuesta elegida
    
     
        guesses.append(selected_option)
    
        if selected_option == answer:                #compara la respuesta elegida con la correcta
            score += 1
            print("¡¡¡Correcto!!!")
        else:
            fails += 1
            print("Incorrecto.")
            print(f"{answer} era la respuesta correcta")
        
    
        print(f"Aciertos: {score} | Fallos: {fails}")
        
        if fails == 5:
            print("\nHas fallado 5 veces. Fin del juego")
            break                                                   # romper el bucle for

        if score == 9:
           print("\n¡Has acertado 9 preguntas! ¡Ganaste!")
           break
    
        
print("\n__________________")

print(f"Tu puntuación final es: {score}")

print(f"Has fallado: {fails}")


    
    
```

```python
# Paso 1: Definir la estructura de datos (El "Backlog") creo un diccionario 
# separo por categorias (historia, ciencias, entretenimiento y cultura general)
# defino qué va a preguntar el programa y cuál es la respuesta correcta 
# para que luego la pueda comparar la IA 
```

```python
# Lista de diccionarios con la base de datos del juego
preguntas_trivia = [
    {
        "categoria": "Ciencias",
        "pregunta": "¿Cuál es el planeta más grande del sistema solar?",
        "opciones": ["Marte", "Júpiter", "Saturno"],
        "respuesta": "Júpiter"
    },
    {
        "categoria": "Historia",
        "pregunta": "¿En qué año comenzó la Segunda Guerra Mundial?",
        "opciones": ["1939", "1945", "1914"],
        "respuesta": "1939"
    },
    {
        "categoria": "Entretenimiento",
        "pregunta": "¿Quién es el protagonista de la saga Indiana Jones?",
        "opciones": ["Brad Pitt", "Harrison Ford", "Tom Cruise"],
        "respuesta": "Harrison Ford"
    },
    {
        "categoria": "Cultura General",
        "pregunta": "¿Qué idioma se habla en Brasil?",
        "opciones": ["Español", "Francés", "Portugués"],
        "respuesta": "Portugués"
    }
]
```

```python
# Prueba rápida: imprimir la primera pregunta para verificar
print(f"Categoría: {preguntas_trivia[0]['categoria']}")
print(f"Pregunta: {preguntas_trivia[0]['pregunta']}")
```

```python
class JuegoTrivia:
    def __init__(self, lista_preguntas):
        # Al iniciar el juego, cargamos las preguntas y ponemos la puntuación a 0
        self.preguntas = lista_preguntas
        self.puntuacion = 0

    def lanzar_pregunta(self, pregunta_dict):
        # Mostramos la categoría y la pregunta [cite: 21]
        print(f"\nCategoría: {pregunta_dict['categoria']}")
        print(f"Pregunta: {pregunta_dict['pregunta']}")
        
        # Mostramos las opciones
        for i, opcion in enumerate(pregunta_dict['opciones']):
            print(f"{i + 1}. {opcion}")
        
        # Pedimos la respuesta al usuario
        respuesta_usuario = input("Escribe tu respuesta: ")
        
        # Validamos si es correcta para permitir avanzar [cite: 22]
        if respuesta_usuario.lower() == pregunta_dict['respuesta'].lower():
            print("¡Correcto! 🎉")
            self.puntuacion += 1
            return True
        else:
            print(f"Incorrecto. La respuesta era: {pregunta_dict['respuesta']} ❌")
            return False

    def jugar(self):
        print("¡Bienvenido al juego de Preguntas y Respuestas!")
        for p in self.preguntas:
            self.lanzar_pregunta(p)
        print(f"\nJuego terminado. Tu puntuación final es: {self.puntuacion}")
```

```python
# probamos que todo funciona
# Creamos el objeto del juego pasando nuestra lista de preguntas
mi_juego = JuegoTrivia(preguntas_trivia)

# Iniciamos la partida
mi_juego.jugar()
```

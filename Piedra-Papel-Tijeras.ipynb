Piedra Papel & Tijera

- Reglas del Juego "Piedra, Papel y Tijera":

    - Dos jugadores eligen una de las tres opciones: "Piedra", "Papel" o "Tijera".

    - Las reglas para determinar al ganador son:
        
        - Piedra vence a Tijera.

        - Tijera vence a Papel.

        - Papel vence a Piedra.

        - El jugador que gana una ronda obtiene un punto.
        
        - El juego continúa hasta que un jugador haya acumulado 3 puntos, convirtiéndose en el ganador.


```python
# El juego se gana con 3 puntos
# El jugador_1 empieza en 0
# El ordenador empieza en 0
# Solo se puede elegir: piedra, papel o tijera
# El juego continúa hasta que un jugador haya acumulado 3 puntos, convirtiéndose en el ganador.


import random

def jugar_piedra_papel_tijera():

    puntos_objetivos = 3
    puntos_jugador = 0
    puntos_ordenador = 0

    opciones = ["piedra", "papel", "tijera", "salir"]

    while puntos_jugador < puntos_objetivos and puntos_ordenador < puntos_objetivos:
        jugador_1 = input("Elige: piedra, papel o tijera (o escribe 'salir'): ").lower()

        if jugador_1 == "salir":
            print("Has salido del juego 👋")
            break

        while jugador_1 not in opciones:
            print ("Tienes que elegir una de las tres opciones que te ofrecí")
            jugador_1 = input ("Elige: piedra, papel o tijera:").lower()    

        computadora= random.choice (opciones)
        print ("Computadora eligió", computadora)

        if jugador_1==computadora:
            print ("Empate")

        elif jugador_1== "piedra" and computadora== "tijera":
            print ("Ganaste")
            puntos_jugador +=1

        elif jugador_1== "papel" and computadora == "piedra":
            print ("Ganaste")
            puntos_jugador += 1 

        elif jugador_1== "tijera" and computadora == "papel":  
            print ("Ganaste")
            puntos_jugador += 1 

        else:
            print ("Perdiste")
            puntos_ordenador += 1

        print("Jugador:", puntos_jugador, "- Computadora:", puntos_ordenador)

        if puntos_jugador== puntos_objetivos:
            print ("¡Ganaste el juego!")
            break

        if puntos_ordenador== puntos_objetivos:
            print ("La computadora ganó el juego")
            break
     
      
```

```python
jugar_piedra_papel_tijera()
```

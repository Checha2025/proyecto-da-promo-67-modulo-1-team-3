```python
import random

preguntas_trivia = [
    {
        "categoria": "Ciencias",
        "pregunta": "¿Cuál es el planeta más grande del sistema solar?",
        "opciones": ["Marte", "Júpiter", "Saturno", "Venus"],
        "respuesta": "Júpiter"
    },
    {
        "categoria": "Historia",
        "pregunta": "¿En qué año comenzó la Segunda Guerra Mundial?",
        "opciones": ["1939", "1945", "1914", "1929"],
        "respuesta": "1939"
    },
    {
        "categoria": "Entretenimiento",
        "pregunta": "¿Quién es el protagonista de la saga Indiana Jones?",
        "opciones": ["Brad Pitt", "Harrison Ford", "Tom Cruise", "Johnny Depp"],
        "respuesta": "Harrison Ford"
    },
    {
        "categoria": "Cultura General",
        "pregunta": "¿Qué idioma se habla en Brasil?",
        "opciones": ["Español", "Francés", "Portugués", "Italiano"],
        "respuesta": "Portugués"
    }
]


class JuegoTrivia:
    def __init__(self, lista_preguntas):
        self.preguntas = lista_preguntas[:]
        random.shuffle(self.preguntas)   # Mezcla el orden de las preguntas
        self.puntuacion = 0
        self.incorrectas = 0
        self.fallos = []

    def lanzar_pregunta(self, pregunta_dict):
        print(f"\nCategoría: {pregunta_dict['categoria']}")
        print(f"Pregunta: {pregunta_dict['pregunta']}")

        opciones = pregunta_dict["opciones"][:]
        random.shuffle(opciones)   # Mezcla el orden de las opciones

        letras = ["A", "B", "C", "D"][:len(opciones)]

        for i, opcion in enumerate(opciones):
            print(f"{letras[i]}. {opcion}")

        respuesta_usuario = input("Elige una opción (A, B, C, D): ").upper()

        if respuesta_usuario not in letras:
            print("Opción no válida ❌")
            return False

        indice = letras.index(respuesta_usuario)
        opcion_elegida = opciones[indice]

        if opcion_elegida.lower() == pregunta_dict["respuesta"].lower():
            print("¡Correcto! 🎉")
            self.puntuacion += 1
            return True
        else:
            print(f"Incorrecto. La respuesta era: {pregunta_dict['respuesta']} ❌")
            self.incorrectas += 1
            
            self.fallos.append({
                "pregunta": pregunta_dict["pregunta"],
                "tu_respuesta": opcion_elegida,
                "respuesta_correcta": pregunta_dict["respuesta"]
            })
            
            return False

    def jugar(self):
        print("¡Bienvenido al juego de Preguntas y Respuestas!")

        for pregunta in self.preguntas:
            self.lanzar_pregunta(pregunta)

        print("\nPuntuación")
        print(f"Correctas: {self.puntuacion}")
        print(f"Incorrectas: {self.incorrectas}")

        if self.fallos:
            print("\n❌ Preguntas falladas:")
            for fallo in self.fallos:
                print(f"- {fallo['pregunta']}")
                print(f"  Tu respuesta: {fallo['tu_respuesta']}")
                print(f"  Correcta: {fallo['respuesta_correcta']}")
        else:
            print("\n🎉 ¡Perfecto! No fallaste ninguna pregunta.")

        print(f"\nPuntuación final: {self.puntuacion}")


mi_juego = JuegoTrivia(preguntas_trivia)
mi_juego.jugar() 
```
